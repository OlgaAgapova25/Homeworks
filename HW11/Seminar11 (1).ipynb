```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
fx = -5*x**2 + 10*x - 150
fx
```

```python
plot(fx) #построить график
```

```python
solve(fx,x) #найти корни
```

```python
solve(fx>0) #вычислить промежутки, на которых функция больше 0
```

```python
solve(fx<0) #вычислить промежутки, на которых функция меньше 0
```

```python
plot(diff(fx))
```

```python
solve(diff(fx)<0) #промежутки, на которых функция убывает
```

```python
solve(diff(fx)>0) #промежутки, на которых функция возрастает
```

```python
solve(diff(fx),x) #вычислить вершину
```
